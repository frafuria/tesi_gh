```python
def identify_cases(y_test, y_pred, threshold=0.5):
    """
    Identify four cases based on the comparison of y_pred and y_test with a threshold.
    Returns a dictionary containing four lists of indices for each case.
    """
    # Calculate the absolute difference between y_pred and y_test
    diff = y_pred - y_test

    # Create masks for the four cases
    mask_case1 = (diff <= - t)
    mask_case2 = (diff < 0) & (diff > - t)
    mask_case3 = (diff >= t)
    mask_case4 = (diff > 0) & (diff < t)

    # Identify the indices for each case
    case1_indices = np.where(mask_case1)
    case2_indices = np.where(mask_case2)
    case3_indices = np.where(mask_case3)
    case4_indices = np.where(mask_case4)

    # Print the percentages of data points in each case
    n_y = len(y_pred)
    print("Case 1 (y_pred < y_test by at least 0.5):", mask_case1.sum()/n_y)
    print("Case 2 (y_pred < y_test by no more than 0.5):", sum(mask_case2)/n_y)
    print("Case 3 (y_pred > y_test by at least 0.5):", sum(mask_case3)/n_y)
    print("Case 4 (y_pred > y_test by no more than 0.5):", sum(mask_case4)/n_y)

    # Create a dictionary to store the indices for each case
    cases = {
        '1': case1_indices,
        '2': case2_indices,
        '3': case3_indices,
        '4': case4_indices,
    }

    return cases
```
