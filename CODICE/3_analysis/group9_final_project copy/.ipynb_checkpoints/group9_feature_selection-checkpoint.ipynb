```python
# IMPORTING STATEMENTS
import pandas as pd
import numpy as np
#import matplotlib.pyplot as plt

import plotly.express as px
```

```python
from sklearn.feature_selection import SelectKBest
```

```python
data = pd.read_excel('default of credit card clients.xls', sheet_name='Data')
data.columns = data.iloc[0,:]
data.drop(index=0, inplace=True)
data.drop(columns='ID', inplace=True)
data = data.astype(int)
data
```

```python
# THIS ONE TURNED OUT TO BE WRONG

data['EDUCATION_encoded'] = ''
data.EDUCATION_encoded = data.EDUCATION_encoded.mask(data.EDUCATION==1, 'graduate')
data.EDUCATION_encoded = data.EDUCATION_encoded.mask(data.EDUCATION==2, 'university')
data.EDUCATION_encoded = data.EDUCATION_encoded.mask(data.EDUCATION==3, 'high school')
data.EDUCATION_encoded = data.EDUCATION_encoded.mask(data.EDUCATION==0, 'other or unknown')
```

```python
data['PUNCTUALITY_AVG'] = data[['PAY_0', 'PAY_2','PAY_3','PAY_4','PAY_5','PAY_6']].mean(axis=1)
data['BILL_AVG'] = data[['BILL_AMT1', 'BILL_AMT2','BILL_AMT3','BILL_AMT4','BILL_AMT5','BILL_AMT6']].mean(axis=1)
data['PAY_AVG'] = data[['PAY_AMT1', 'PAY_AMT2','PAY_AMT3','PAY_AMT4','PAY_AMT5','PAY_AMT6']].mean(axis=1)
```

```python
data.columns
```

```python
y = data['default payment next month']
```

```python
all_labels = ['LIMIT_BAL', 'SEX', 'EDUCATION', 'MARRIAGE', 'AGE', 'PAY_0', 'PAY_2',
       'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6', 'BILL_AMT1', 'BILL_AMT2',
       'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6', 'PAY_AMT1',
       'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6',
        'PUNCTUALITY_AVG', 'BILL_AVG', 'PAY_AVG']
X_all_labels = data[all_labels]

```

```python
only_avgs = ['LIMIT_BAL', 'SEX', 'EDUCATION', 'MARRIAGE', 'AGE', 'PUNCTUALITY_AVG', 'BILL_AVG', 'PAY_AVG']
X_only_avgs = data[only_avgs]
```

```python
from sklearn.preprocessing import StandardScaler
ss= StandardScaler()

X_all_labels_trans = ss.fit_transform(X_all_labels)
X_only_avgs_trans = ss.fit_transform(X_only_avgs)
```

```python
results = pd.DataFrame(index=all_labels, columns=['5K_all', '10K_all','5K_avgs',
                                                 'pvalue5_all', 'pvalue5_avgs', 'pvalue1_all', 'pvalue1_avgs'])
```

```python
# SELECT 5 BEST FROM X_ALL_LABELS

selector = SelectKBest(k=5)
selector.fit_transform(X_all_labels,y)
selected = selector.get_feature_names_out(input_features=all_labels)

for var in all_labels:
    if var in selected:
        results.loc[var, '5K_all']=1
    else:
        results.loc[var, '5K_all']=0
```

```python
# SELECT 10 BEST FROM X_ALL_LABELS

selector = SelectKBest(k=10)
selector.fit_transform(X_all_labels,y)
selected = selector.get_feature_names_out(input_features=all_labels)

for var in all_labels:
    if var in selected:
        results.loc[var, '10K_all']=1
    else:
        results.loc[var, '10K_all']=0
```

```python
# SELECT 5 BEST FROM X_ONLY_AVG

selector = SelectKBest(k=5)
selector.fit_transform(X_only_avgs,y)
selected = selector.get_feature_names_out(input_features=only_avgs)

for var in all_labels:
    if var not in only_avgs:
        results.loc[var, '5K_avgs']=2
    elif var in selected:
        results.loc[var, '5K_avgs']=1
    else:
        results.loc[var, '5K_avgs']=0
```

```python
# 0.001 p-value on ALL_LABELS
selector = SelectFpr(alpha=0.001)
X_new = selector.fit_transform(X_all_labels,y)
selected = selector.get_feature_names_out(input_features= all_labels)

for var in all_labels:
    if var in selected:
        results.loc[var, 'pvalue1_all']=1
    else:
        results.loc[var, 'pvalue1_all']=0
```

```python
# 0.005 p-value on ONLY_AVG
selector = SelectFpr(alpha=0.005)
X_new = selector.fit_transform(X_only_avgs,y)
selected = selector.get_feature_names_out(input_features=only_avgs)

for var in all_labels:
    if var not in only_avgs:
        results.loc[var, 'pvalue5_avgs']=2
    elif var in selected:
        results.loc[var, 'pvalue5_avgs']=1
    else:
        results.loc[var, 'pvalue5_avgs']=0
```

```python
def is_true(s):
    if s == 1:
        color = 'lightgreen'
    elif s == 0:
        color = 'red'
    else:
        color = 'white'
    
    return 'background-color: %s' % color 

results.style.applymap(is_true)
```
