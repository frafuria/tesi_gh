# Import Data

```python
# Import true test data
data_test = pd.read_csv('../../0_data/test.csv', index_col=0).reset_index(drop=True)
```

```python
# select random sample from test data
counterf = data_test.sample(n=750, replace=False, random_state=7, axis=None, ignore_index=False) 
counterf.describe()
```

```python
# create counterfactual cases

#counterf.drop(['Minority_Min', 'Minority_Missing', 'Minority_White'], axis=1, inplace=True)
counterf1, counterf2, counterf3 = counterf.copy(deep=True), counterf.copy(deep=True), counterf.copy(deep=True)

counterf1['Minority_Min'], counterf2['Minority_Min'], counterf3['Minority_Min'] = 1, 0, 0
counterf1['Minority_Missing'], counterf2['Minority_Missing'], counterf3['Minority_Missing'] = 0, 1, 0
counterf1['Minority_White'], counterf2['Minority_White'], counterf3['Minority_White'] = 0, 0, 1
#y_true = data_test.pop('InterestRate')
```

```python
# DF FINALE 
counterf = pd.concat([counterf1, counterf2, counterf3])
counterf.to_csv('TEST_DATA.csv')
```
